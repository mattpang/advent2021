```python
#tally up the bracket type
brascore = dict()
brascore['['] = 1
brascore[']'] = -1

score = dict()
score[')'] =3 
score[']']=57
score['}']=1197
score['>']=25137
total=0

pair=dict()
pair['('] = ')'
pair['{'] = '}'
pair['['] = ']'
pair['<'] = '>'

keeps=[]
for i,line in enumerate(lines):
    mem=[]
    keeper =True
    #print(len(line))
    for pos,c in enumerate(line):
        if c in pair.keys():
            mem.append(c)
        elif pair[mem[-1]] != c:
            #print(line,c,pos)
            total+=score[c]
            keeper=False
            break
        elif pair[mem[-1]] == c:
            mem.pop()
    if keeper:
        keeps.append(line)
#total
```

```python
import copy
#recurivsly remove all the full brackets, then when there's only left brackets, we stop:

# ): 1 point.
# ]: 2 points.
# }: 3 points.
# >: 4 points.

score['(']=1
score['[']=2
score['{']=3
score['<']=4

total = []
for keep in keeps:
    k = copy.deepcopy(keep)
    lastlen=30
    while True:
        k = k.replace('[]','')
        k= k.replace('()','')
        k= k.replace('<>','')
        k= k.replace('{}','')

        if lastlen==len(k):
            break
        lastlen=len(k)
    #score things up. just the reversed order. 
    k = k[::-1]
    
    tally=0
    for s in k:
        tally*=5
        tally+=score[s]
    print(k,tally)
    total.append(tally)

sorted(total)[len(total)//2]
```
