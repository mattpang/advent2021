```python
import copy
xv,yv = 50,50
allhits=[]

for myv in range(-1000,1000,1):
    for mxv in range(0,1000,1):
        xpos = 0
        ypos = 0
        t=0
        maxy = 0 
        hit=False
        pos_store = []
        xv = copy.deepcopy(mxv)
        yv = copy.deepcopy(myv)
        
        while True:            
            t+=1
            xpos+=xv
            ypos+=yv
            xv-=1
            if xv<0:
                xv=0
            yv-=1
            #print(xpos,ypos)
            pos_store.append((xpos,ypos))
            if (xa[0]<=xpos<=xa[1]) and (ya[0]<=ypos<=ya[1]):
                print('hit!')
                print([t,mxv,myv,xv,yv,maxy])
                hit=True
                allhits.append([t,mxv,myv,maxy])#,pos_store])
                break

            if (ypos < min(ya)) or (xpos > max(xa)):
                #print('Beyond range')
                break
            if ypos > maxy:
                maxy=ypos

```
