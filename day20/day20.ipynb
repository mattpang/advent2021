```python
fmt=lines[0]
e=[]
for c in fmt:
    if c=='#':
        e.append(1)
    elif c=='.':
        e.append(0)
    else:
        print('error') 
len(e)   
```

```python
graw = lines[2:-1]
store = []
for row in graw:
    l = []
    for c in row:
        if c=='#':
            l.append(1)
        elif c=='.':
            l.append(0)
        else:
            print('error') 
    store.append(l)

```

```python
def scorer(blist):
    ans= 0
    for i,b in enumerate(blist[::-1]):
        if b == 1:
            ans+=2**i
    return ans

filler = [e[0], e[511] if e[0] else e[0]]
ngrid = grid

for i in range(50):
    #print('filler = ',filler[(i+1)%2])
    ngrid = np.pad(ngrid,2,constant_values=filler[(i+1)%2])
    newgrid = np.ones_like(ngrid)*filler[i%2]
    for j in range(1,len(ngrid)-1):
        for i in range(1,len(ngrid[0])-1):
            tally=[]
            tally+=list(ngrid[j-1,i-1:i+2])
            tally+=list(ngrid[j,i-1:i+2])
            tally+=list(ngrid[j+1,i-1:i+2])
            #print(i,j,scorer(tally))
            newgrid[j,i] = e[scorer(tally)]
    
    ngrid=newgrid
    
ngrid.sum()
```
