```python
f=open('input.txt','r')
lines = f.read().split('\n')[:-1]
```

```python
lines = '''NNCB

CH -> B
HH -> N
CB -> H
NH -> C
HB -> C
HC -> B
HN -> C
NN -> C
BH -> H
NC -> B
NB -> B
BN -> B
BB -> N
BC -> B
CC -> N
CN -> C'''.split('\n')
```

```python
start=lines[0]
```

```python
rules=lines[2:]
```

```python
seq = lines[0]
for step in range(0,10):
    print(step)
    store='x'
    pos = 0
    for double in [seq[x:x+2] for x in range(0,len(seq)+1)]:
        #print(double)
        for rule in rules:
            result,newelement = rule.split(' -> ')
            insert = result[0]+newelement+result[1]
            #seq = seq.replace(result,insert)
            if double == result:
                if store[-1]==double[0]:
                    store+=insert[1:]
                else:
                    store+=insert
                break
            #    store+=double[0]
    seq = store[1:]
    #print(step,seq)
    
```

```python
# do it as a histogram
pairs = Counter()
elements=Counter()
seq = lines[0]

for c in seq:
    elements[c]+=1
print(elements)

for double in [seq[x:x+2] for x in range(0,len(seq)+1)]:
    if len(double)==2:
        pairs[double] +=1
        
rules = dict()
for r in lines[2:]:
    p,v = r.split(' -> ')
    rules[p] = v
        
for step in range(0,40): 
    pairlist = list(pairs.keys())

    new = Counter()
    for pair, num in pairs.items():
        if pair in rules:
            a, b = pair
            c = rules[pair]
            new[a + c] += num
            new[c + b] += num
            elements[c] += num
        else:
            new[pair] = num
    pairs = new
                
    #print(pairs)
    #print(elements)
ranked = elements.most_common()
ranked[0][1] - ranked[-1][1]
```
