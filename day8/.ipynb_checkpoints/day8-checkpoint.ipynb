```python
f = open('input.txt','r')
lines = f.read().split('\n')[:-1]
```

```python
#part 1 is just count the letters
from collections import Counter
tally =0
for line in lines:
    wires, displayed = line.split('|')
    segs = displayed.strip().split(' ')
    for seg in segs:
        if len(seg) == 2:
            tally +=1
        elif len(seg) == 7:
            tally +=1

        elif len(seg) == 4:
            tally +=1

        elif len(seg) == 3:
            tally +=1


tally
```

```python
#all 8 would be 8, 2 would be 1, 4 is 3 etc
# want to map the random seg orders to the real one. 
# # 
#  aaaa   
# b    c  
# b    c  
#  ....   
# e    f  
# e    f  
#  gggg

# what is in 7 and not 1, is the top edge. 
# what is in 0 and not 8 is the middle stroke. 
# what is in 9 and not 8 is the lower left stroke. 
# but 0,9 are too similar to tell 
# what is in 4 is top left edge

total = 0 

from collections import defaultdict
for line in lines:
    
    nums = set(range(0,10))
    wires, displayed = line.split('|')
    print(line.split('|'))
    segs = [''.join(x) for x in list(map(sorted,wires.strip().split(' ')))]
    shown = [''.join(x) for x in list(map(sorted,displayed.strip().split(' ')))]
    possible = dict()
    #matches = defaultdict(set)
    matches = dict()
    possible069=[]
    lettermap=dict()
    len5=set()
    fivesegs = set()
    for seg in segs:
        if len(seg) == 2:
            matches[seg] = 1
            nums.remove(1)
            possible[1] = seg
        elif len(seg) == 7:
            matches[seg] = 8
            nums.remove(8)
            possible[8] = seg
            
        elif len(seg) == 4:
            matches[seg] = 4
            nums.remove(4)
            possible[4] = seg
        elif len(seg) == 3:
            matches[seg] = 7
            nums.remove(7)
            possible[7] = seg
            seven = seg
        elif len(seg) == 6:
            possible069.append(seg)

    
    
    lettermap['b'] = list(set(possible[7])-set(possible[1]))[0]
    fivesegs=set(possible[4])-set(possible[1])
#    lettermap['d']


    
    for seg in segs:
        if len(seg) == 6:
            if (seg not in matches.keys()) and (set(seven).issubset(seg)):
                #print('9 matched:',seg)
                matches[seg] = 9
                
    for seg in segs:
        if len(seg) == 5:
            if (seg not in matches.keys()) and (set(seven).issubset(seg)):
                #print('3 matched:',seg)
                matches[seg] = 3
                
    p = set(possible[4])-set(possible[1])

    seenin5=Counter()
    for seg in segs:
        if len(seg) == 5:
            #print(seg)
            for pv in p:
                if pv in seg:
                    seenin5[pv]+=1
                
            if p.issubset(set(seg)):
                #print(seg,'has middle seg in')
                #print(p)
                matches[seg] = 5
    
    try:
        assert list(seenin5.values())[0]==3
    except AssertionError:
        print(seenin5)
        
    lettermap['d']= seenin5.most_common()[0][0]
    
    # now can get 0: 
    for seg in segs: 
        if len(seg) == 6:
            #print('poss zero',seg)
            if lettermap['d'] not in seg:
                matches[seg] = 0
    

                
    for seg in segs:
        if (len(seg) == 5) and (seg not in matches.keys()):
            #print('found 2',seg)
            matches[seg] = 2
    
    for seg in segs:
        if (seg not in matches.keys()):
            #print('found 6',seg)
            matches[seg] = 6
    

    amounts = 0
    for unit,s in zip([1000,100,10,1],shown):
        amounts += unit*matches[s]
    total+=amounts
    print(amounts)
```
