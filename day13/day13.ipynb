```python
lines = '''6,10
0,14
9,10
0,3
10,4
4,11
6,0
6,12
4,1
0,13
10,12
3,4
3,0
8,4
1,10
2,14
8,10
9,0

fold along y=7
fold along x=5'''.split('\n')
```

```python
f=open('input.txt','r')
lines=f.read().split('\n')[:-1]
```

```python
folds = []
dots = []
maxy=0
maxx=0
for line in lines:
    if line.startswith('fold'): 
        folds.append(line)
    else:
        try:
            x,y = map(int,line.split(','))
            dots.append((x,y))
            if x>maxx:
                maxx=x
            if y>maxy:
                maxy=y
        except:
            pass
        
dots
```

```python
maxx,maxy
```

```python
import numpy as np

zshape = (maxy+1,maxx+1)
z = np.zeros(shape=zshape)
z
```

```python
for x,y in dots:
    z[y,x] = 1
z
```

```python
newz = z.copy()
for fold in folds:
    val = int(fold.split('=')[-1])
    if 'x' in fold:
        #need to split grid in half, along
        z = newz[:,:val]+np.fliplr(newz[:,val+1:])
    elif 'y' in fold:
        #take bottom half, flip and add to the top half
        z = newz[:val,:]+np.flipud(newz[val+1:,:])
    newz = z.copy()
    #print((newz>0).sum())
    #break
```

```python
import matplotlib.pyplot as plt

plt.imshow(newz>0)
```
